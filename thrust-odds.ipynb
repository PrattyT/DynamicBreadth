```python
import math
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Optional - RUN THIS ONCE A DAY TO GET THE LATEST DATA. Takes a while to run.
closingPrices = pd.read_csv("latestClose.csv")
closingPrices.tail()

tickers = closingPrices.columns.to_list()[1::]
def getSearchRange(start, end):
    return closingPrices[closingPrices['Date'].between(start, end)]

#drop todays data
closingPrices = getSearchRange('2021-01-01', '2023-05-07')

# Get the closing prices for the last 100 days and calculate the average daily change magnitude
def getAverageDailyChange(ticker):
    last100Days = closingPrices[ticker].tail(100)
    return last100Days.pct_change().abs().mean()

averageMove = {}
for ticker in tickers:
    averageMove[ticker] = getAverageDailyChange(ticker)

```

```python
closingPrices.tail()
```

```python
# add 1 to each value 
averageMultiplier = {}
for ticker in tickers:
    averageMultiplier[ticker] = averageMove[ticker] + 1


```

```python
#check if todays price multiplied by multiplier is a new 64 day high
def checkIfNewHigh(ticker, multiplier):
    last64Days = closingPrices[ticker].tail(64)
    return last64Days.max() < closingPrices[ticker].tail(1).values[0] * multiplier

output = []
for ticker in tickers:
    print(ticker, averageMultiplier[ticker])
    if checkIfNewHigh(ticker, averageMultiplier[ticker]):
        output.append(ticker)


```
